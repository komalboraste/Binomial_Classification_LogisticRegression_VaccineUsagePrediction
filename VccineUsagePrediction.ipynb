```python
df.h1n1_vaccine=df.h1n1_vaccine.astype('object')
```

```python
numeric=df.select_dtypes(include=['float64'])
```

```python
numeric.columns
```

```python
import numpy as np
```

```python
for i in numeric:
    med=numeric[i].median(axis=0)
    numeric[i].replace(np.nan,med,inplace=True)
```

```python
numeric.isna().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#Correlation plot
corr=numeric.corr()
plt.subplots(figsize=(20,9))
sns.heatmap(corr,annot=True)
```

```python
categorical=df.select_dtypes(include='object')
```

```python
categorical.columns
```

```python
for c in categorical:
    null=categorical[c].value_counts().index[0]
    categorical[c].replace(np.nan,'None',inplace=True)
```

```python
categorical.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Label encoding categorical variables
for c in categorical:
    lbl=LabelEncoder()
    lbl.fit(list(categorical[c].values))
    categorical[c]=lbl.transform(list(categorical[c].values))
```

```python
categorical.head()
```

```python
new_df=pd.concat([numeric,categorical],axis=1)
```

```python
new_df.columns
```

```python
new_df.isna().sum()
```

```python
y=new_df['h1n1_vaccine']
```

```python
y.head()
```

```python
X=new_df[['h1n1_worry', 'h1n1_awareness', 'antiviral_medication',
       'contact_avoidance', 'bought_face_mask', 'wash_hands_frequently',
       'avoid_large_gatherings', 'reduced_outside_home_cont',
       'avoid_touch_face', 'dr_recc_h1n1_vacc', 'dr_recc_seasonal_vacc',
       'chronic_medic_condition', 'cont_child_undr_6_mnths',
       'is_health_worker', 'has_health_insur', 'is_h1n1_vacc_effective',
       'is_h1n1_risky', 'sick_from_h1n1_vacc', 'is_seas_vacc_effective',
       'is_seas_risky', 'sick_from_seas_vacc', 'no_of_adults',
       'no_of_children', 'age_bracket', 'qualification', 'race', 'sex',
       'income_level', 'marital_status', 'housing_status', 'employment',
       'census_msa']]
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr2=LogisticRegression(max_iter=1000)
```

```python
#fit the model
lr.fit(X_train,y_train)
```

```python
lr2.fit(X_train,y_train)
```

```python
#making predictions using data
y_pred=lr.predict(X_test)
```
